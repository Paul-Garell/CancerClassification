```python

def filter(file,gene,freq,seq,start):

    df = pd.read_csv(file, sep="\t") 

    #Filtering out columns
    start_index_1 = 0
    end_index_1 = 3
    df.drop(df.columns[start_index_1:end_index_1], axis=1, inplace=True)
    
    start_index_2 = 1
    end_index_2 = 5
    df.drop(df.columns[start_index_2:end_index_2], axis=1, inplace=True) 

    start_index_3 = 2
    end_index_3 = 7
    df.drop(df.columns[start_index_3:end_index_3], axis=1, inplace=True)
    
    start_index_4 = 7
    end_index_4 = df.shape[1]
    df.drop(df.columns[start_index_4:end_index_4], axis=1, inplace=True)


    #Adding new columns
    df['Gene Name'] = gene
    df['Frequency'] = freq
    df['Reference'] = seq
    df['Ref Start'] = start

    return df
```
